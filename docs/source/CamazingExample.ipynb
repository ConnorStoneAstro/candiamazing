# Candiamazing Example Usage

You can write jupyter notebooks and have them appear in your website!

```python
import candiamazing as ca
import matplotlib.pyplot as plt
import numpy as np
```

```python
Phot = ca.BrightnessConverter(zeropoint=21.4)

print(Phot.mag_to_flux(20.0))  # Example usage
```

```python
f = np.logspace(0, 3, 100)
m = Phot.flux_to_mag(f)

plt.plot(f, m, linewidth=2, c="k")
plt.xlabel("Flux (arbitrary units)")
plt.ylabel("Magnitude")
plt.gca().invert_yaxis()
plt.title("Flux to Magnitude Conversion")
plt.grid()
plt.show()
```
